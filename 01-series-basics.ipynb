```python
import pandas as pd
import numpy as np
```

```python
# Series in pandas
s = pd.Series([23, 24, 25, 26])
print(s)
print(type(s))

# Indexing
print("\nYou can also print by index just like normal arrays")
print(s[0])    # 23
print(s[2])    # 25

print("\nThe following gives the index range")
print(s.index)     # all labels
```

# Custom Indexing

```python
# Custom Indexing
print("\nInstead of conventional index, you can use your own")
s2 = pd.Series([23, 24, 25, 26], index = ["Adam", "Eve", "Charlie", "Bob"])
print(s2["Eve"])    # 24
print(s2["Bob"])    # 26
```

# Vectorized Operations

```python
print("\nSupports vectorized operations just like numpy arrays")
s1 = pd.Series([1, 2, 3])
s2 = pd.Series([4, 5, 6])

print(s1 + s2)
```

# Mutable Values but immutable size

```python
print("\nSize is immutable")
s = pd.Series([1, 2, 3, 4, 5])
s[0] = 100

print(s)
```

```python
print("\nWhen we change an element or drop it, it creates a new series")
changed_s = s.drop(1) # Takes label as parameter
print("\nNow")
print(changed_s)
print("\nBefore")
print(s)
```
