# Filtering Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("globalAirQuality.csv")
```

```python
df[df["aqi"] > 100] # prints rows where aqi>100
```

```python
df[df['aqi']>100][['city', 'aqi']]
# Prints the city and aqi where aqi > 100
```

```python
df[(df["aqi"] > 100) & (df["timestamp"] == "2025-11-04 18:25:17.554219")]
# prints the rows where aqi > 100 and the specific timestamp
```

```python
# you can also store the specific condition in a variable to further apply conditions to it.
aqi_data = df[(df['aqi']>100) & (df['timestamp'] == '2025-11-04 18:25:17.554219')][['city','aqi']]
# Selects the city and aqi column with aqi > 100 and specific timestamp
```

```python
aqi_data.loc[360] # prints the 360 row with the condition specified
aqi_data.iloc[1]
```

# Query Methods

```python
# Query - returns a copy, not a view
# we write the condition as a string
df.query("aqi > 100")

# other operators can also be passed as a query
df.query("aqi > 100 & timestamp == '2025-11-04 18:25:17.554219'").head()
# using head to return the top 5 only
```

```python
# We could also use a logic like f-string in python
# using variables inside query method
my_country = "US"
df.query("country == @my_country") # add @ before the variable
```

```python
# chained comparisons
df["timestamp"] = pd.to_datetime(df["timestamp"])
df["date"] = df["timestamp"].dt.date
selected_date = pd.to_datetime("2025-11-04").date()

df.query("country == @my_country & date == @selected_date & aqi > 100")
```

```python
# Use .copy() to avoid confusion of view or copy
countries = df[["country", "aqi"]].copy() # copies the data
countries.loc[0:4, "country"] = "India" # changes the new data country column from 1 to 4 row to India

print(countries)
```
