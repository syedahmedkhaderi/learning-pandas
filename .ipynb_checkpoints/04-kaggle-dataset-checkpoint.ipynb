```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("globalAirQuality.csv")
```

```python
df.describe()
```

# Selecting Data

```python
# Select columns
df["country"]
df[["city", "aqi"]]
```

```python
# Select rows (by label)
df.loc[0]             # 1st row (by label)
df.loc[0:3]           # row0 to row3 - both inclusive in loc
```

```python
# Select rows (by index)
df.iloc[0]            # 1st row 
df.iloc[4:7]          # 1st row, Ending index is exclusive.
```

```python
# Select rows & columns
df.loc[0, "country"]  # 0th row & specific column
df.iloc[0, 2]         # prints 1th row & 3rd column (by position)
```

```python
df.iloc[0, 1:5]
df.loc[0, ["country", "aqi"]] # Prints country and aqi of 0th row
```

```python
df.loc[0:5, ["country", "city", "latitude"]]   # end is inclusive
df.iloc[0:5, 1:3] # Prints 0-5 rows and 1-3 columns
```

```python
# Select single scalar value
df.at[0, "city"]
df.iat[0, 2] # Treat it as matrix, element at 0th row and 2nd column.
```
